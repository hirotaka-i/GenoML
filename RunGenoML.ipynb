```python
import os
import subprocess
from os import chdir
import io
import sys

tmp = '/lscratch/' + os.environ['SLURM_JOB_ID'] + '/GenoML'
if os.path.exists(tmp):
    print(tmp, 'exists')
else:
    os.mkdir(tmp)
    print('Make', tmp)

chdir(tmp)
```

```python
%%bash
wget https://github.com/ipdgc/GenoML-Brief-Intro/raw/master/exampleData.zip
unzip exampleData.zip  
```

```python
%%bash
source /data/LNG/iwakih2/genoML/conda/etc/profile.d/conda.sh
conda activate genoMLproject
genoml-train \
 --geno-prefix=exampleData/training \
 --pheno-file=exampleData/training.pheno \
 --gwas-file=exampleData/example_GWAS.txt \
 --cov-file=exampleData/training.cov \
 --addit-file=./exampleData/training.addit \
 --model-file=exampleModel
```

```python
%%bash
source /data/LNG/iwakih2/genoML/conda/etc/profile.d/conda.sh
conda activate genoMLproject
genoml-inference \
 --model-file=exampleModel \
 --valid-dir=exampleOUTPUT \
 --valid-geno-prefix=exampleData/validation \
 --valid-pheno-file=exampleData/validation.pheno \
 --valid-cov-file=exampleData/validation.cov \
 --valid-addit-file=exampleData/validation.addit
```
